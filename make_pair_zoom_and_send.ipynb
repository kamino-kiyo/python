```python
import glob
import os
import re
import rstr
import shutil

zoom_suffix = "Zoom"
send_suffix = "MobileL"

# 各テーマフォルダ下の_savefiledebugフォルダ
debug_path = (r"D:\puri\theme\20Sum_Happy\Theme\_savefiledebug").replace("\\", "/")
print(debug_path)

# リネーム移動先（この下にテーマ名フォルダを作ってそこに移す）
dst_path = (r"C:\Users\0FR115005\TmpData\Zoom_Send").replace("\\", "/")
print(dst_path)
```

```python
print("start")

# リネーム移動先にテーマ名フォルダを作る
theme_name = re.findall("theme/(.*)/Theme", debug_path)[0]
zoom_dir = dst_path + "/" + theme_name + "/" + zoom_suffix
send_dir = dst_path + "/" + theme_name + "/" + send_suffix
os.makedirs(zoom_dir, exist_ok=True)
os.makedirs(send_dir, exist_ok=True)


# debugフォルダ内の画像でZoom-Mobileのペアを作る
# Zoom1, Zoom2, ... Zoom6, Mobile1, Mobile2, ... Mobile6, Zoom1, ...の順で並んでるファイルを
# Zoom1とMobile1が対応するようにファイル名リネームした上でZoomとMobileを別々のフォルダに分ける処理

# ZoomのリストとMobileのリストに分ける
zoom_list = []
send_list = []
for img_path in glob.glob(debug_path + "/*.bmp"):
    if zoom_suffix in img_path:
        zoom_list.append(img_path)
    elif send_suffix in img_path:
        send_list.append(img_path)

# ZoomとMobileの数が違ったらエラーにしちゃう（微妙？）
if len(zoom_list) != len(send_list):
    print(zoom_suffix + "と" + send_suffix + "の数が違うのでやめます")
    exit()

# ZoomとMobileのファイル名を対応させてからフォルダ移動
for (zoom_path, send_path) in zip(zoom_list, send_list):
    zoom_num = re.findall(zoom_suffix + "(.)", zoom_path)[0]
    send_num = re.findall(send_suffix + "(.)", send_path)[0]
    
    # 同じKeep番号じゃなかったら中断（しなくてもいい？？）
    if zoom_num != send_num:
        print("Keepが違うのでやめます")
        continue
    
    # ランダムファイル名生成
    # https://qiita.com/ryasuna/items/083ee0dc5e2db484e458
    rand = rstr.xeger(r'^[0-9a-zA-Z0-9]{14}')
    new_zoom_path = zoom_dir + "/" + rand + ".bmp"
    new_send_path = send_dir + "/" + rand + ".bmp"
    
    # リネーム＆フォルダ移動
    print(shutil.move(zoom_path, new_zoom_path))
    print(shutil.move(send_path, new_send_path))
    print("-----------------------------------")
    
print("end")
```
