```python
from pathlib import Path
import csv
import shutil
import xml.etree.ElementTree as ET

# LabelingToolForML用 LabelInfo.csvの内容で画像をフォルダ分けする

# 【入力】Setting.xmlとLabelInfo.csvが置いてあるフォルダのパスを指定する
src_path = Path(r"D:\puri\theme\20Sum_Happy\Theme\_savefiledebug\CAOLABO営業ロケテVROSAKA2台_1296x全Shot_メイク仕分け")

# -------------------------------------
# Setting.xmlからラベル一覧の取得
tree = ET.parse(src_path.as_posix() + "/Setting.xml")
root = tree.getroot()

# ついでにフォルダも作っておく
image_dir = Path(src_path.as_posix() + "/check")
image_dir.mkdir(exist_ok=True)

# ラベル名とアイテム名のリストを作る
label_itemtuple_list = []
for label in root.iter("Label"):
    item_list = []
    label_name = label.get("name")

    label_dir = Path(image_dir.as_posix() + "/" + label_name)
    label_dir.mkdir(exist_ok=True)
    
    for item in label.iter("Item"):
        item_name = item.get("name")
        item_dir = Path(label_dir.as_posix() + "/" + item_name)
        item_dir.mkdir(exist_ok=True)
                
        item_list.append(item_name)    
    
    label_itemtuple_list.append((label_name, item_list))

# -------------------------------------
# LabelInfo.csv
# ファイルをラベルごとに当てはまるフォルダにコピーしていく
with open(src_path.as_posix() + "/LabelInfo.csv") as info:
    reader = csv.reader(info)
    for row in reader:
        image_name = row[0]
        items = row[1:]
        for i in range(len(items)):
            # Item名が"-"だったら未分類なので無視する
            if items[i] == "-":
                continue
            image_path = src_path.as_posix() + "/" + image_name.replace("\\", "/")
            dst_path = image_dir.as_posix() + "/" + label_itemtuple_list[i][0] + "/" + items[i]
            shutil.copy(image_path, dst_path)
 
            




```
