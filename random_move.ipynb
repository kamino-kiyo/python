```python
# in1フォルダ内のファイル数に合わせて
# in2フォルダ内のファイルをランダムに選んで outフォルダに移動させる
import glob
import os
import random
import shutil

base = r"D:\docomo5G\PoC4\20220117_服\機種C(ネコカノ)".replace("\\", "/")
in1 = base + "/制服"
in2 = base + "/制服以外"
out = base + "/制服以外/枚数"

os.makedirs(out, exist_ok=True)

# num1 = len([name for name in os.listdir(in1) if os.path.isfile(os.path.join(in1, name))])
num1 = 718
num2 = len([name for name in os.listdir(in2) if os.path.isfile(os.path.join(in2, name))])

if num1 == num2:
    print("やる必要なし")
if num1 > num2:
    print("パス指定ミスってそう")
else:
    list = glob.glob(in2 + "/*")
    rand = random.sample(range(num2), k=num1)
    for i in rand:
        path = shutil.move(list[i], out)
        print(path)

print("end")
```
